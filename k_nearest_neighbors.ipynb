```python
import matplotlib as plt
import pandas as pd
import seaborn as sns
```

```python
url = "http://dibresources.jcbose.ac.in/ssaha4/pulmopred/public/training.csv.txt"
ds = pd.read_csv(url)

ds = ds[~ds['Diagnosis'].isin(['DPLD', 'OSA', 'Sarcodiosis','Chest Pain'])]
ds = ds.drop(columns = ['Label'])
X_without_diagnosis = ds.drop('Diagnosis', axis = 1)

X = X_without_diagnosis
Y = ds['Diagnosis']

sns.scatterplot(data=X, x=X_without_diagnosis.columns[0], y=Y, hue=Y)
#sns.scatterplot(data=X, x='FVC Pre-BD Value', y='FVC Pre-BD Pred', hue=hue)

#sns.scatterplot(X = X_without_diagnosis, Y=ds['Diagnosis'], hue=ds['Diagnosis'])
```
